```python
from because.probability import prob

from because.synth import read_data
from because.visualization import viz

r = read_data.Reader('/home/arun/HPCC-covid19/LLCP/llcpA.csv')
ds = r.read()

ps = prob.ProbSpace(ds)
print('ps.N = ', ps.N)
print('ps.fieldList = ', ps.fieldList)
print('ds shape = ',len(ps.fieldList))  

# 0.9490824254306831
```

## Hypothesis 1: Diabetes and Age are positively correlated.

Probability of diabetes increases with the age. 

Let's see what is the relaionship between age and height.

## Diabetes | Age

```python
%matplotlib inline
viz.show(probspace=ps, targetSpec=[('diabetes','yes')], condSpec=['age'], controlFor=[], filtSpec=[], gtype='pdf', power=1, enhance=True)
```

```python
%matplotlib inline
viz.show(probspace=ps, targetSpec=[('diabetes')], condSpec=['age'], controlFor=[], filtSpec=[], gtype='exp', power=1, enhance=True)
```
