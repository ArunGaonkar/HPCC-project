```python
from because.probability import prob

from because.synth import read_data
from because.visualization import viz

r = read_data.Reader('/home/arun/HPCC-covid19/LLCP/llcpA.csv')
ds = r.read()

ps = prob.ProbSpace(ds)
print('ps.N = ', ps.N)
print('ps.fieldList = ', ps.fieldList)
print('ds shape = ',len(ps.fieldList))  
```

```python
ps.P(('depression', 'yes'),('state', 'NY'))
```

```python
print('State Vals = ', ps.getValues('state'))
print('Weight Vals = ', ps.getValues('weight'))
print('Ed Vals = ', ps.getValues('education'))
print('Income Vals = ', ps.getValues('income'))
print('Asthma Vals = ', ps.getValues('asthma'))
```

```python
%matplotlib inline
viz.show(probspace=ps, targetSpec=['height'], condSpec=[], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['height'], condSpec=[], filtSpec = [('gender', 'male')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['height'], condSpec=[], filtSpec = [('gender', 'female')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['weight'], condSpec=[], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['weight'], condSpec=[], filtSpec = [('gender', 'male')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['weight'], condSpec=[], filtSpec = [('gender', 'female')], gtype='pdf', power=1, enhance=True)
```

```python
d = ps.distr('weight')
print(len(d.ToHistTuple()))
```

```python
%matplotlib widget
viz.show(probspace=ps, targetSpec=['genhealth'], condSpec=['age'], controlFor=[], filtSpec=[('state', 'CA')], gtype='pdf', power=1, enhance=True)
```

```python
print(ps.P(('gender','male'),[('weight', 180), ('height', 67),('age',65)]), 
        ps.P(('gender','female'),[('weight', 130), ('height', 62)]))

print(ps.P(('gender','male'),[('weight', 180), ('height', 72),('age',65)]), 
        ps.P(('gender','female'),[('weight', 130), ('height', 64)]))
```

```python
# expectation value of weight
print(ps.E( targetSpecs='weight', givenSpecs=[('age',62),('height',70), ('gender','male')]))
print(ps.E( targetSpecs='weight', givenSpecs=[('age',62),('height',70), ('gender','female')]))
```
